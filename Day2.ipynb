```python
import numpy as np

a=np.array([1,2,3,4,5])
b=np.arange(1,6)
c=np.zeros((3))
d=np.ones((3,2))

print("Array a:",a)
print("Array b:",b)
print("2*3 Zeros:",c)
print("3*2 Ones:",d)
```

```python
print("First element of a:",a[0])
print("Last two elements of a:",a[-2:])
```

```python
print("a+2:",a+2)
print("a*3:",a*3)
```

```python
print("Square of a:", np.square(a))
print("Mean of a:", np.mean(a))
print("Standard deviation:", np.std(a))
```

```python

matrix =np.arange(1,10).reshape((3,3))
print("Reshaped 3*3 matrix:")
matrix
```

```python
print("Transpose:")
matrix.T
```

```python
print("Elements > 5:", matrix[matrix > 5])
```

```python
import pandas as pd

data = {
    'Name':['Alice','Bob','Charlie','David','Eva'],
    'Age':[25,30,35,40,22],
    'Department':['Sales','Engineering','HR','Engineering','Sales'],
    'Salery':[50000,80000,40000,90000,45000]
}
df=pd.DataFrame(data)
df
```

```python
 print("First 3 rows:")
 df.head(3)
```

```python
print("Column names:")
df.columns
```

```python
print("Data types:")
df.dtypes
```

```python
print("Describe numerical columns:")
df.describe()
```

```python
print("People with salary >50000:")
df[df['Salary'] >50000]
```

```python
print("Sorted by Age:")
df.sort_values(by='Age')
```

```python
grouped = df.groupby('Department')['Salary'].mean()
print("Average Salary by Department:")
grouped
```
